```python
import pandas as pd                                 #Pandas is used to analyze data
import smtplib                                      #For sending mail 
from email.mime.multipart import MIMEMultipart      #For adding attachments or providing alternate version of same content
from email.mime.text import MIMEText                #For text emails
from email.mime.base import MIMEBase                #Base class for MIME
from email import encoders

def py_mail(SUBJECT, BODY, TO, FROM, FROMNAME):
    """With this function we send out our HTML email"""

    # Create message container - the correct MIME type is multipart/alternative here!
    MESSAGE = MIMEMultipart('alternative')
    MESSAGE['subject'] = SUBJECT
    MESSAGE['To'] = TO
    MESSAGE['From'] = FROMNAME
    MESSAGE.preamble = """
Your mail reader does not support the report format.
Please visit us online!"""

    # Record the MIME type text/html.
    HTML_BODY = MIMEText(BODY, 'html')

    # Company Profile (Attachment)
    attach_file_name = 'About_Us_-_Hackveda_Limited.pdf'
    attach_file = open(attach_file_name, 'rb') # Open the file as binary mode
    payload = MIMEBase('application', 'octate-stream')
    payload.set_payload((attach_file).read())
    encoders.encode_base64(payload) #encode the attachment
    #add payload header with filename
    payload.add_header('Content-Disposition', 'attachment', filename=attach_file_name)
    MESSAGE.attach(payload) # Enable/Disable Email Attachment
    
    # Devanshu Shukla Profile
    attach_file_name = 'Devanshu_Shukla_Brief_Profile.pdf'
    attach_file = open(attach_file_name, 'rb') # Open the file as binary mode
    payload = MIMEBase('application', 'octate-stream')
    payload.set_payload((attach_file).read())
    encoders.encode_base64(payload) #encode the attachment
    #add payload header with filename
    payload.add_header('Content-Disposition', 'attachment', filename=attach_file_name)
    MESSAGE.attach(payload) # Enable/Disable Email Attachment
    
    # Devanshu Shukla Certifications
    attach_file_name = 'Certifications_Credentials.pdf'
    attach_file = open(attach_file_name, 'rb') # Open the file as binary mode
    payload = MIMEBase('application', 'octate-stream')
    payload.set_payload((attach_file).read())
    encoders.encode_base64(payload) #encode the attachment
    #add payload header with filename
    payload.add_header('Content-Disposition', 'attachment', filename=attach_file_name)
    MESSAGE.attach(payload) # Enable/Disable Email Attachment
    
    # Attach parts into message container.
    # According to RFC 2046, the last part of a multipart message, in this case
    # the HTML message, is best and preferred.
    MESSAGE.attach(HTML_BODY)

    # The actual sending of the e-mail
    server = smtplib.SMTP('smtp.gmail.com:587')

    # Print debugging output when testing
    if __name__ == "__main__":
        server.set_debuglevel(0)

    # Credentials (if needed) for sending the mail
    password = "hynwfmorahkfpqep"

    server.starttls()
    server.login(FROM,password)
    server.sendmail(FROMNAME, [TO], MESSAGE.as_string())
    server.quit()
```
