```python
# Queue implementation
class Queue:
  def __init__(self, size):
    self.q = [None] * size
    self.capacity = size
    self.front = 0
    self.rear = -1
    self.count = 0
    
  def pop(self):
    if self.isEmpty():
      print("Terminate")
      exit(1)
    print("Removing element: ", self.q[self.front])
    self.front = (self.front + 1) % self.capacity
    self.count = self.count - 1

  def append(self, value):
    if self.isFull():
      print("Overflow")
      exit(1)
    print("Inserting: ", value)
    self.rear = (self.rear + 1) % self.capacity
    self.q[self.rear] = value
    self.count = self.count + 1

  def peek(self):
    if self.isEmpty():
      print("Queue empty")
      exit(1)
    
    def size(self):
      return self.count

    def isEmpty(self):
      return self.size() == 0
    
    def isFull(self):
      return self.size() == self.capacity

if __name__ == "__main__":
  cust_queue = Queue(10)
  print("Size of the queue is", cust_queue.size())
  cust_queue.append(10)
  cust_queue.append(20)
  cust_queue.append(30)
  cust_queue.append(40)
  print("Size of the queue is", cust_queue.size())
  cust_queue.pop()
  if cust_queue.isEmpty():
    print("Empty Queue")
  else:
    print("Non Empty Queue")
```
