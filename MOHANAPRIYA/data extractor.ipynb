```python
table=browser.find_element_by_id("tablaDatos")
```

```python
#search the elements within the table
thead=table.find_element_by_tag_name("tr")
thead
```

```python
th_list=thead.find_elements_by_tag_name("th")
th_list
```

```python
for th in th_list:
    heading=th.text
    print(heading)
```

```python
tbody=table.find_element_by_tag_name('tbody')
tbody
```

```python
#search for tr tags within tbody
tr_list=tbody.find_elements_by_tag_name("tr")
tr_list   
```

```python
import pandas as pd

event_name_list=[]
start_date_list=[]
organizer_name_list=[]
event_address_list=[]
end_date_list=[]
my_data={}


# iterate over each table row(tr) and extract data
#tr_list=tr_list[1:]
for tr in tr_list:
    try:
        td_list=tr.find_elements_by_tag_name("td")
        
        start_date=td_list[0].text
        event_name=td_list[1].text
        event_type=td_list[2].text
        
        event_location_data=td_list[3]
        location=event_location_data.find_elements_by_tag_name("span")
        organizer_name=location[1].text
        event_address=location[2].text
        
        # event_organizer_name=event_location_data.find_element_by_xpath("//span[@itemprop='name']").text
        # event_address=event_location_data.find_element_by_xpath("//span[@itemprop='address']").text
        
        end_date=td_list[4].text
        
        if start_date == "" or event_name == "":
            pass
        else:
            print("Event Name:",event_name,"Start Date:",start_date,"Organizer Name:",organizer_name,
                 "Event Address:",event_address,"Event Ends:",end_date)
            print("\n\n")
            
            event_name_list.append(event_name)
            start_date_list.append(start_date)
            organizer_name_list.append(organizer_name)
            event_address_list.append(event_address)
            end_date_list.append(end_date)
    except Exception as e:
        print("Exception",e)
        pass
result_dict={"Event Name":event_name_list, "Start Date":start_date_list,"Organizer Name":organizer_name_list,
             "Event Address":event_address_list, "Events Ends":end_date_list}
#result_dict={"Event Name:",tuple(event_name_list)}
my_data=pd.DataFrame(data=result_dict)
```
