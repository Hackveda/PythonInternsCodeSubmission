```python
# Fibonacci series
# The sum of two elements defines the next
a , b = 0 , 1
while a < 10:
  print(a)
  a , b = b , a + b
```

```python
# Simple Grade System Application
# 91-100 = A, 81-90 = B, 71-80 = C, Below or equal 70 = fail

# take input from the user
marks = int(input("Enter your marks: "))

#Conditional Flow Implementation is IF
if marks >= 91:
   print("Grade A")
elif marks >= 81:
    print("Grade B")
elif marks >= 71:
    print("Grade C")
else:
  print("Failed Score")
```

```python
# Switch Implementation of IF
# ATM Machine Application
card = input("Enter Card: ")
print("""
Select your option number

1. Banking
2. Bank Statement
3. PIN Change
4. Fast Withdrawl
""")

option = int(input("Enter your option number: "))

#Switch Implementation
if option == 1:
  print("Welcome to Banking Mode")
elif option == 2:
  print("Print Bank Statement")
elif option == 3:
  print("PIN Change Menu")
elif option == 4:
  print("Fast Withdrawl")
else:
  print("Please check the correct option5")
```

```python
# for Statements in Python
# Railway Reservation Application

queue = ["Tej", "Amit", "Shubham", "Neetej"]

#Start making tickets for this queue using FOR statement
for traveller_name in queue:
  print("Make Ticket for", traveller_name)
  print("Ticket Issued for", traveller_name)
print("All travellers ticket made")
```

```python
# Add all the items in a basket starting from character B
items = ["Banana", "Apple", "Burger", "Juice", "Orange", "Butter", "Brinjal", "Coffee", "Milk", "Bread"]

basket = []
for i in items:
  if i[0] == 'B':
    basket.append(i)
basket
```
