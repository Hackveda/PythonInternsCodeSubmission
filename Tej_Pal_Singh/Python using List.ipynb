```python
# Concarenation in lists
list1 = ["Tej Pal Singh", 29, 5.7]
list2 = ["Neetu Singh", 30, 5.2]
con_result = list1 + list2
con_result
```

```python
# Lists are mutable, their content can be changed
index = 0
for item in con_result:
  print(index, item)
  index = index + 1
```

```python
#change value in list
con_result[1] = 30
con_result
```

```python
#add new items into the list
con_result.append('9871')
con_result
```

```python
#update slice in the list
list2 = ["Monika", 31, 5.3]
con_result[1:6] = list2
con_result
```

```python
#remove items from the list
con_result[1:5] = []
con_result
```

```python
#calculate the size of list
len(con_result)
```
