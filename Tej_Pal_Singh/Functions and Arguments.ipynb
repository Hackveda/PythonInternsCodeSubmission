```python
# Defining the arguments list
# Create an application to make a team and add players
def ipl_team(teamName, *players):
  "Create a new IPL Team"
  player_list = []
  for player in players:
    player_list.append(player)
  print("IPL Team", teamName, "Player List:", player_list)
```

```python
ipl_team("Mumbai Indians", "Sachin tendulkar", "Rohit Sharma", "Hardik Pandya", "Polard")
```
