```python
list1 = [1,2,3,4,5] #Create a List
list1
```

```python
# Append an item on the list
list1.append(6) #Time complexity O(1)
list1
```

```python
# Extend Operation
list2 =[7,8,9] # K Elements
list1.extend(list2) # O(k) Time Complexity
list1
```

```python
# Calculating the time complaxity of extend operation
list1 = [1,2,3,4,5] # N elements
list1
list2 = [6,7,8,9] # K elements
list2

# Hoe list2 will be extended within list1
for element in list2:
  list1.append(element)

list1
```

```python
# Insert Operation
list1.insert(0,10) # Add at head of list = O(1)
list1.insert(7,11) # Add 11 at index 10 of list = O(N)
list1.insert(11,12) # Add to the tail of the list = O(1)
list1
```

```python
# Remove Operation
list1.remove(10)
list1
```

```python
# Pop Operation
list1.pop(0)
list1
```

```python
# clear Operation
list1.clear() #O(1)
list1
```
